# Computation Graph

```python
import torch
import torch.nn as nn
from torchviz import make_dot
```

**Nota :** we also need `pixi global install graphviz` to have `dot` available.

```python
model = nn.Sequential()
model.add_module('W0', nn.Linear(8, 16))
model.add_module('tanh', nn.Tanh())
model.add_module('W1', nn.Linear(16, 1))

x = torch.randn(1, 8)
y = model(x)

make_dot(y.mean(), params=dict(model.named_parameters()))
```

```python
x = torch.tensor(1.0, requires_grad=True)
y = torch.tensor(2.0, requires_grad=True)
w = x + 2.0 * y + 1.0
z = w*w + torch.sin(x)
```

```python
make_dot(z, params={"x": x, "y": y})
```

Interesting! Probably not exactly what I would like, since the operations that are listed are the gradient computations (and not the forward value computation), but interesting nonetheless.
